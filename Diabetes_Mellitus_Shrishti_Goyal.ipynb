```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = pd.read_csv('/content/sample_data/diabetes.csv')
```

```
data.head()
```

```
print(data.dtypes)
print(data.columns)
```

```
data.info()
```

There is no missing value.

```
data.describe()
```

Minimum value for BMI, Glucose, Skin Thickness and Blood Pressure is zero which is not possible.

Replacing zero value with mean

```
data.loc[:,'Glucose'].replace(0,data.loc[:,'Glucose'].mean())
data.loc[:,'BloodPressure'].replace(0,data.loc[:,'BloodPressure'].mean())
data.loc[:,'SkinThickness'].replace(0,data.loc[:,'SkinThickness'].mean())
data.loc[:,'BMI'].replace(0,data.loc[:,'BMI'].mean())
```

```
data.describe()
```

**Pairwise Correlation**


```
correlation = data.corr()
print(correlation)
sns.heatmap(correlation)
```

Age & Pregnancies and BMI & SkinThickness have moderate positive linear relationship. Insulin and glucose has slightly positive linear relationship.

```
sns.pairplot(data)
```

From scatter plots, only BMI & SkinThickness and Pregnancies & Age seem to have positive linear relationships. Glucose and Insulin have slightly positive relationship.
There are no non-linear relationships.
